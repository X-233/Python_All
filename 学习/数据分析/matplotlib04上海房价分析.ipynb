```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['axes.unicode_minus'] = False
```

```python
shanghai_data = eval(open('data_1.text', 'r', encoding='utf-8').read())
fig, ax = plt.subplots(nrows=6, ncols=2, constrained_layout=True, figsize=(20, 40))
# ax[0, 0].plot(shanghai_data[0]['data_mon'].keys(), shanghai_data[0]['data_mon'].values(), label=shanghai_data[0]['year'])
for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        ax[i, j].plot(shanghai_data[i*(ax.shape[1])+j]['data_mon'].keys(), shanghai_data[i*(ax.shape[1])+j]['data_mon'].values(), label=str(shanghai_data[i*(ax.shape[1])+j]['year'])+'年', color='black')
        ax[i, j].set_yticks(range(10000, 50000, 10000))
        ax[i, j].legend(fontsize=15)
        ax[i, j].grid(True, linestyle="--", alpha=0.5)
plt.savefig('房价分析.png')
```

```python
years = [i['year'] for i in shanghai_data]
sums = [sum(i['data_mon'].values()) / len(shanghai_data) / 10000 for i in shanghai_data]
plt.figure(figsize=(20, 8), constrained_layout=True)
plt.plot(years, sums, color='yellow', label='2011-2022年房价趋势')
plt.legend(fontsize=15)
plt.grid(True, linestyle="--", alpha=0.3)
plt.yticks(range(1, 6), fontsize=15)
plt.xticks(years, fontsize=15)
plt.xlabel("Years", fontsize=20, loc="right")
plt.ylabel("Price/万￥", fontsize=20, loc="top")
print(len(years))
```

```python
plt.figure(figsize=(20, 8), constrained_layout=True)
plt.barh(years, sums, label='2011-2022')
plt.legend(fontsize=15)
plt.yticks(years, fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel("Price/万￥", fontsize=20, loc='right')
```
