```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['axes.unicode_minus'] = False
```

```python
movies = {
    "": [2.01, 4.59, 7.99, 11.83, 16],
    "": [3.19, 5.08, 6.73, 8.10, 9.35],
    "": [4.07, 6.92, 9.30, 11.29, 13.03],
    "": [2.72, 3.79, 4.45, 4.83, 5.11],
    "": [0.56, 0.74, 0.83, 0.88, 0.92],
    "": [0.66, 0.95, 1.10, 1.17, 1.23],
    "": [0.58, 0.81, 0.94, 1.01, 1.07],
    "": [1.13, 1.96, 2.73, 3.42, 4.05]
}

name = movies.keys()
data = list(movies.values())
everyday = []
every = []
x2 = range(len(name))
width = 0.80
s_width = width/len(data[0])

plt.figure(figsize=(20, 8))
for j in range(len(data[0])):
    for i in range(len(data)):
        every.append(data[i][j])
    everyday.append(every)
    every = []
for i in everyday:
    xs = [j - width//2 + (everyday.index(i) -2)*s_width for j in x2]
    plt.bar(xs, i, width=s_width, label=f'{everyday.index(i)+1}')
    for ticks, x in zip(i, xs):
        plt.annotate(ticks, xy=(x, ticks), ha='center', xytext=(x, ticks+0.1), fontsize=12)

plt.xticks(x2, name, fontsize=15)
plt.title('')
plt.ylabel('/', fontsize=15)
plt.legend()

print()
```
