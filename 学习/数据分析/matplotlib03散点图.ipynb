```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['axes.unicode_minus'] = False
```

```python
from random import random

plt.figure(figsize=(20, 8))

data_x = [random() for i in range(10)]
data_y = [random() for j in range(10)]

plt.scatter(data_x, data_y)
```

```python
size = [100*random() for i in range(10)]
color = ['red', 'green', 'blue', 'black', 'yellow', 'purple', 'black', 'yellow', 'purple', 'red']
plt.scatter(data_x, data_y, s=size, c=color, alpha=0.5, marker='H')
```

```python
data_x = [random() for i in range(10)]
data_y = [random() for j in range(10)]

data_x1 = [random() for i in range(10)]
data_y1 = [random() for j in range(10)]

plt.scatter(data_x, data_y, marker='X', label='')
plt.scatter(data_x1, data_y1, marker='P', label='')
plt.legend()
```
