```python
import matplotlib.pyplot as plt
import random
import re
plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = [i for i in range(20)]
y_1 = [random.uniform(2, 10) for i in range(20)]
y_2 = [random.uniform(4, 14) for i in range(20)]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)
axes[0].plot(x, y_1, label='1', color='black',)
axes[1].plot(x, y_2, label='2', color='red',)

axes[0].legend()
axes[1].legend()

x_1 = [i for i in range(20)]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_1[::5])
axes[0].set_yticks(range(-10, 30)[::5])
axes[0].set_xlabel('', fontsize=20)

x_2 = [i for i in range(20)]
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_2[::5])
axes[1].set_yticks(range(-10, 30)[::5])
axes[1].set_xlabel('', fontsize=20)

axes[0].grid()
axes[1].grid()
```

```python
fix, axes = plt.subplots(nrows=1, ncols=2, figsize=(200, 80), dpi=100)
axes[0].barh(x, y_1, label='y_1', color='black')
axes[1].barh(x, y_2, label='y_2', color='green')
```
